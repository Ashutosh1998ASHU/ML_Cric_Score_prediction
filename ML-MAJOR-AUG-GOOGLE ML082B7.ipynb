# MATCH SCORE PREDICTOR


```python
import numpy as np
import pandas as pd
import sklearn as sk
import os
df = pd.read_csv("ipl2017.csv", index_col = 0)
df.head()
```

```python
os.getcwd()
```

#  Handling missing values

```python
df.info()
```

# Removing unnecessary columns

```python
df = df.drop(["date","venue"],axis=1)
df.head()
```

#  Converting non numeric columns into numeric columns

```python
df = pd.get_dummies(df,columns = ["bat_team","bowl_team","batsman","bowler"])
df.head()
```

```python
df.info()
```

```python
df.shape
```

# Extracting

```python
y = df["total"]
X = df.drop("total", axis = 1)
```

```python
X.shape
```

```python
X.to_csv("NewDataset2.csv", encoding = "utf-8")
```

# Data splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Scalling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(X_train,y_train)
```

# Score checking

```python
forest.score(X_test,y_test)
```

# New dataset prediction 

```python
NewDataset = pd.read_csv("NewDataset2.csv", index_col = 0)
NewDataset.head()
```

```python
NewDataset.shape
```

```python
forest.predict(NewDataset)
```
